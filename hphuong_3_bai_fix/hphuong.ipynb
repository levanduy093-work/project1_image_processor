```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
```

Corner line

```python
img = cv2.imread('img/white_bear.jpg')
plt.imshow(img[:,:,::-1])
plt.axis('off')
plt.show()
print(img.shape)
```

```python
def drawCornerLine(img,width):
    h,w = img.shape[:2]
    start_x = w//5
    for row in range(h):
        for col in range(w):
            if(col <= (start_x + width - row) and col >= (start_x - width - row)):
                img[row, col,:] = 0
    return img
img_copy = img.copy()
img_copy = drawCornerLine(img_copy,50)
plt.imshow(img_copy[:,:,::-1])
plt.axis('off')
plt.show()
```

Gradient

```python
#Gradient
#To mt mng 2 chiu vi kch thc 256x256
img_arr = np.zeros((256,256),dtype=np.uint8)
img_arr.shape
#Vit hm thay i sc  t trn xung di
def Gradient_down(img):
    g_img = img.copy()
    h,w = img_arr.shape
    for i in range(h):
        g_img[i,:] = 255-i
    return g_img
g_img = Gradient_down(img_arr)
plt.imshow(g_img, cmap='gray')
plt.axis("off")
plt.show()
```

Letter B

```python
#To nh
img = np.ones((400,400),dtype=np.uint8)*255

def draw_B(img, width, size_text):
    h,w = img.shape
    hText, wText = size_text
    #Kim tra kch thc ch v nh
    if(h<hText or w<wText):
        raise ValueError("Size text ln hn kch thc nh") 
    #Tm im bt u v ch
    start_x = (w - wText ) // 2
    start_y = (h - hText) // 2
    #V phn dc ca ch B
    img[start_y:start_y + hText, start_x:start_x + width] = 0
    #V 3 ng ngang
    pNgang = (hText)//5
    for i in [0, 2, 4]:  # Cc phn ngang th 0, 2, 4
        y = start_y + i * pNgang
        img[y:y + pNgang, start_x + width:start_x + wText - width] = 0
    #V 2 nt dc bn phi
    for i in [1, 3]:  # Cc phn ngang th 1, 3
        y = start_y + i * pNgang
        img[y-width//2:y + pNgang+width//2, start_x + wText - width:start_x + wText] = 0
    return img
img = draw_B(img,50,(300,200))
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.show()
```
